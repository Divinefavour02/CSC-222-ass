```python
data.columns = data.columns.str.strip()  # Remove leading/trailing spaces
```

```python
# to display summary statistics
data.describe
```

```python
# to rename fund to financial gatherings
data.rename(columns = {'Fund':'Financial Gatherings'})
```

```python
# to drop a column
data.drop(columns=['Cert'])
```

```python
# converting CLOSING DATE column to_datetime data type
data['Closing Date'] = pd.to_datetime(data['Closing Date'], format='%d-%b-%y')
```

```python
# using value counts
data['City'].value_counts()
```

```python
# Example for grouping
data.groupby(['City']).agg({'Fund': 'sum', 'Cert': 'mean'})
```

```python
# matplotlib visualisations
import matplotlib.pyplot as plt
```

```python
# using matplotlib pie chart
plt.pie(data['Fund'], labels = data['City'])
plt.show()
```

```python
# using matplotlib graphs
plt.plot(data['City'], data['Fund'])
plt.show()
```

```python
# using matplotlib barchart
plt.bar(data['City'], data['Fund'])
plt.show()
```

```python
# using matplotlib scatterplots
plt.scatter(data['City'], data['Fund'])
plt.show()
```

```python
# seaborn visualizations
import seaborn as sns
```

```python
# seaborn bar plot
city_funds = data.groupby('City')['Fund'].sum().reset_index().sort_values('Fund', ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x='City', y='Fund', data=city_funds)
plt.xticks(rotation=90)
plt.title('Total Fund by City')
plt.tight_layout()
plt.show()
```

```python
# seaborn boxplot
plt.figure(figsize=(14, 6))
sns.boxplot(x='City', y='Fund', data=data)
plt.xticks(rotation=90)
plt.title('Fund Distribution by City')
plt.tight_layout()
plt.show()
```

```python
# seaborn histogram
plt.figure(figsize=(8, 5))
sns.histplot(data['Fund'], bins=15, kde=True, color='skyblue')
plt.title('Distribution of Fund Values')
plt.xlabel('Fund')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# seaborn count plot
plt.figure(figsize=(12, 6))
sns.countplot(x='City', data=data, order=data['City'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Number of Institutions per City')
plt.tight_layout()
plt.show()
```
